```python
%matplotlib notebook
import os 
import csv
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
```

```python
runway_use_file = "SFO_Air_Carrier_Runway_Use.csv"
```

```python
#read csv
runway_use_df = pd.read_csv(runway_use_file)
runway_use_df.head()
```

```python
#check number of values are in each column
runway_use_df.count()
```

```python
#check data type for each column
runway_use_df.dtypes
```

```python
#convert year and month to strings
runway_use_df['Year'] = runway_use_df['Year'].astype(str)
runway_use_df['Month'] = runway_use_df['Month'].astype(str)

runway_use_df.dtypes
```

```python
#filter data for only 2008 to 2018
runway_use_df = runway_use_df.loc[(runway_use_df['Year'] >= '2008') & 
                                  (runway_use_df['Year'] <= '2018'),:]

runway_use_df['Year'].unique()
```

```python
runway_use_df
```

```python
#find total count 
runway_use_df['01L/R'] = runway_use_df['01L/R Departure'] + runway_use_df['01L/R Arrival'] 
runway_use_df['10L/R'] = runway_use_df['10L/R Departure'] + runway_use_df['10L/R Arrival'] 
runway_use_df['19L/R'] = runway_use_df['19L/R Departure'] + runway_use_df['19L/R Arrival'] 
runway_use_df['28L/R'] = runway_use_df['28L/R Departure'] + runway_use_df['28L/R Arrival'] 

runway_use_df = runway_use_df[['Year', '01L/R', '10L/R', '19L/R', '28L/R']]

runway_use_df
```

```python
#group dataframe by Year
runway_use_df = runway_use_df.groupby(['Year']).sum()
runway_use_df
```

```python
#reset index
runway_use_df.reset_index(level=0, inplace=True)
runway_use_df
```

```python
# create line graph
plt.plot('Year', '01L/R', data=runway_use_df, marker='o', markerfacecolor='blue', markersize=2, color='blue', linewidth=1.5)
plt.plot('Year', '10L/R', data=runway_use_df, marker='o', markerfacecolor='yellow', markersize=2, color='yellow', linewidth=1.5)
plt.plot('Year', '19L/R', data=runway_use_df, marker='o', markerfacecolor='green', markersize=2, color='green', linewidth=1.5)
plt.plot('Year', '28L/R', data=runway_use_df, marker='o', markerfacecolor='red', markersize=2, color='red', linewidth=1.5)

plt.legend(loc="upper left")
plt.xlabel('Year')
plt.ylabel('Total # of Arrivals and Deptartures')
plt.xticks(rotation=45)
plt.title('SFO Runway Usage Annual Trends')
plt.ylim((0,300000))
plt.tight_layout()
plt.show()
```

```python
# create bar chart
runway_use_df.plot(x="Year", y=["01L/R", "10L/R", "19L/R", "28L/R"],
                   kind="bar", width=0.5, align="center", stacked = True)


plt.legend(loc="best")
plt.title('SFO Runway Usage')
plt.xlabel('Year')
plt.ylabel('Total Number of Flights')
plt.xticks(rotation=45)
plt.ylim((0,500000))
plt.tight_layout()
plt.show()

#save figure
plt.savefig("./SFO_Runway_Usage_Bar_Chart")
```
